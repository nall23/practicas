## Practica#2: Estadistica descriptiva

Es la rama de la estadística que se encarga de recopilar, organizar, presentar y analizar datos con el objetivo de describir las características principales de un conjunto de datos. En otras palabras, nos ayuda a entender "qué dicen los números"

# Medidas de Tendencia Central:

Estas medidas nos indican hacia dónde se concentran los datos, es decir, cuál es el valor "típico" o representativo de un conjunto de datos. Las más comunes son:

**Media (promedio):** Se calcula sumando todos los valores y dividiéndolos por el número total de datos. Es la medida más utilizada, pero puede ser afectada por valores extremos.
**Mediana:** Es el valor que ocupa la posición central cuando los datos están ordenados de menor a mayor. No se ve afectada por valores extremos.
Moda: Es el valor que más se repite en el conjunto de datos.

# Medidas de Dispersión:

Estas medidas nos indican qué tan dispersos están los datos con respecto a la medida de tendencia central. Es decir, cuán diferentes son los valores entre sí. Las más comunes son:

**Rango:** Es la diferencia entre el valor máximo y el valor mínimo.
**Varianza:** Es el promedio de las desviaciones al cuadrado de cada dato respecto a la media.
**Desviación estándar:** Es la raíz cuadrada de la varianza. Es una medida más intuitiva que la varianza, ya que está en las mismas unidades que los datos originales.

```python
#importamos libreria pandas
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\Acer\Desktop\trabajos facu\7mo semestre\AYDR_Tareas_MineriaDeDatos\csv\steam_limpio.csv'
, encoding='latin-1') #recordar poner la direccion del csv steam_limpio csv
```

```python
#imprimimos para ver los datos
df
```

```python
# Calcular algunas estadísticas descriptivas
print("Resumen estadístico:")
print(df.describe())
```

```python
# Calcular la media de una columna específica(en este caso 'english')
media_columna_english = df['english'].mean()
print("Media de english:", media_columna_english)
```

```python
# Calcular la mediana de otra columna (en este caso 'english')
mediana_columna_english = df['english'].median()
print("Mediana de english:", mediana_columna_english)
```

```python
# Crear un histograma para visualizar la distribución de una columna (en este caso de las reseñas positivas)
import matplotlib.pyplot as plt
plt.hist(df['release_date'], bins=50)
plt.title("Histograma del lanzamiento de los juegos")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Agrupar por desarrollador y el promedio de tiempo de jugadores en sus juegos
jugadores_por_desarrollador = df.groupby('publisher')['average_playtime'].sum()

# Visualizar los 10 desarrolladores con mayor numero de tiempo promedio de juego por parte de sus jugadores
jugadores_por_desarrollador.nlargest(10).plot(kind='bar')
plt.title('Top 10 Compañias por tiempo promedio de jugadores activos')
plt.xlabel('Compañia')
plt.ylabel('Tiempo promedio de juego')
plt.show()
```

```python
# Crear un boxplot para comparar el precio de juegos por genero
sns.boxplot(x='genres', y='price', data=df)
plt.title('Comparación de Precios por Categoría')
plt.show()
#no se aprecian las categorias por el exceso de texto,pero lo importante es ver la grafica jaja
```

```python
sns.boxplot(y='price', x='platforms', data=df)
plt.title("Comparacion de Precios por Plataforma donde se es compatible", fontsize=16)
plt.xlabel("Plataforma", fontsize=25)
plt.ylabel("Precio", fontsize=25)
plt.xticks(rotation=90)
plt.show()
```
