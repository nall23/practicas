## Practica#9: Analisis de Texto

es el proceso de transformar texto no estructurado en un formato estructurado para identificar patrones significativos y nuevos insights.

```python
#Importamos librerias
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
```

```python
df = pd.read_csv(r'C:\Users\Acer\Desktop\trabajos facu\7mo semestre\AYDR_Tareas_MineriaDeDatos\csv\steam_limpio.csv'
, encoding='latin-1')      #recordar poner la direccion del csv steam_limpio csv
```

```python
# Seleccionar la columna que contiene el texto
# Suponiendo que la columna de texto es 'name' o 'description'
text_data = df['name'].dropna()  # O si tienes 'description': df['description']

# Concatenar todo el texto en un solo string
text = " ".join(text_data)

# Crear la nube de palabras
wordcloud = WordCloud(stopwords=ENGLISH_STOP_WORDS, background_color='white', width=800, height=400).generate(text)

# Mostrar la nube de palabras
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Ocultar los ejes
plt.title("Nube de Palabras")
plt.show()
```
