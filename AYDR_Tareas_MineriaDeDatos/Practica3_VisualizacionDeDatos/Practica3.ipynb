## Practica #4: Visualizacion de Datos

la visualización de datos es la representación gráfica de datos e información. Utiliza gráficos, diagramas y otras herramientas visuales para facilitar la comprensión de patrones, tendencias y relaciones en los datos.

### Utilidad:
-Facilita la interpretación rápida de grandes volúmenes de datos.

-Ayuda a identificar patrones, anomalías y correlaciones.

-Es crucial para tomar decisiones basadas en datos.

-Mejora la comunicación al presentar información compleja de manera accesible.

## Técnicas de visualización:
**Gráficos de barras:** Comparan valores de diferentes categorías.

**Histogramas:** Muestran distribuciones de frecuencias.

**Diagramas circulares (Pie Charts):** Representan proporciones dentro de un conjunto.

**Diagramas de dispersión (Scatter Plots):** Muestran relaciones entre dos variables.

**Diagramas de caja (Box Plots):** Identifican distribuciones y valores atípicos.

**Mapas de calor (Heatmaps):** Visualizan patrones en matrices o tablas de datos.

```python
#importamos librerias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Acer\Desktop\trabajos facu\7mo semestre\AYDR_Tareas_MineriaDeDatos\csv\steam_limpio.csv'
, encoding='latin-1')      #recordar poner la direccion del csv steam_limpio csv
```

```python
# Configuracion de estilo para gráficos
sns.set(style="darkgrid")
```

```python
# Crear una lista de columnas numéricas y categóricas para iterar
numeric_columns = ['positive_ratings', 'negative_ratings', 'average_playtime', 'median_playtime', 'price']
categorical_columns = ['platforms', 'categories', 'genres', 'developer', 'publisher']
```

**Diagrama circular:** Representa la proporcion de valores en columnas categoricas.

```python
# 1. Diagrama circular
for col in categorical_columns[:5]:  # Es un ejemplo utilizando las variables categoricas
    counts = df[col].value_counts().head(5)
    plt.figure(figsize=(8, 8))
    counts.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f"Distribución de {col}")
    plt.ylabel("")  # Ocultar etiqueta del eje Y
    plt.show()
```

**Histograma:** Muestra la distribucion de columnas numericas.

```python
# 2. Histograma
for col in numeric_columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[col], kde=True, bins=10)
    plt.title(f"Histograma de {col}")
    plt.xlabel(col)
    plt.ylabel("Frecuencia")
    plt.show()
```

**Diagrama de caja:** Visualiza la dispersion y los valores atipicos de columnas numericas.

```python
# 3. Diagrama de caja
for col in numeric_columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df, y=col)
    plt.title(f"Diagrama de caja de {col}")
    plt.ylabel(col)
    plt.show()
```

**Grafico de lineas:** Representa el cambio en valores numericos a lo largo de los indices.

```python
# 4. Grafico de lineas
for col in numeric_columns[:5]:
    plt.figure(figsize=(10, 6))
    plt.plot(df.index, df[col], label=col)
    plt.title(f"Gráfico de líneas de {col}")
    plt.xlabel("Indice")
    plt.ylabel(col)
    plt.legend()
    plt.show()
```

**Diagrama de dispersion:** Analiza la relacion entre dos columnas numericas

```python
# 5. Diagrama de dispersion
for i, col1 in enumerate(numeric_columns):
    for col2 in numeric_columns[i+1:]:
        plt.figure(figsize=(8, 6))
        sns.scatterplot(data=df, x=col1, y=col2, alpha=0.6)
        plt.title(f"Diagrama de dispersión: {col1} vs {col2}")
        plt.xlabel(col1)
        plt.ylabel(col2)
        plt.show()
```
