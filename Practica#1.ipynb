## Practica#1: Data Cleaning 

La limpieza de datos, o data cleaning, es un proceso fundamental para garantizar la calidad de los datos que se utilizan en el an√°lisis de datos y en la toma de decisiones de una empresa

```python
import pandas as pd
df = pd.read_csv(r'C:\Users\Acer\Desktop\trabajos facu\7mo semestre\AYDR_Tareas_MineriaDeDatos\csv\steam.csv'
                 , encoding='latin-1') #FAVOR DE PONER DIRECCION ESPECIFICA DEL ARCHIVO
print(df)
```

# Ahora veeremos un poco del proceso con codigo de python

```python
# Eliminar filas con valores faltantes en la columna "appid"
df = df.dropna(subset=['appid'])

df
```

```python
# Eliminar duplicados
df = df.drop_duplicates()

df
```

```python
# Veemos el tipo de dato que es cada variable
print(df.dtypes)
```

```python
# Convertir la columna "edad" a tipo entero
df['required_age'] = df['required_age'].astype(int)

# Mostrar el DataFrame limpio
df
```

```python
# Eliminar columnas inecesarias
df = df.drop(columns=['required_age','steamspy_tags','achievements','pricexd' ]
             , inplace=False)
df
```

```python
# Guardar el DataFrame como un nuevo archivo CSV
df.to_csv("steam_limpio.csv", index=False)
```
